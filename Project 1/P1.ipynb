Show that for any such matrix, one has $\|A\|_2=\rho(A)$ (where $\rho(A)$ is the spectral radius of $A$).</font>


By Theorem 1.2 in the textbook, we have
$$\|A\|_2=\sigma_1(A), \tag{1}$$
where $\sigma_1(A)$ is the largest singular value of A. The singular value can also be expressed as the square roots of non-negative eigenvalues of $A^H A$, that is, $\sigma(A)=\lambda(A^HA)$.

Since $A$ is a normal matrix with a unitary diagonalization $U\Lambda U^H$, we get

$$
\sqrt{A^H A}=\sqrt{U\Lambda^2 U^H}=U\sqrt{\Lambda^2}U^H=U|\Lambda|U^H$$

Furthermore, since $\Lambda$ is a diagonal matrix, its eigenvalues are simply the diagonals $\lambda_i$.
$$
\therefore \sigma_i(A) = |\lambda_i| \\$$
$$
\implies \sigma_1(A) = \max_i |\lambda_i| \tag{2}
$$

The spectral radius of a matrix, by definition is $\max_i {|\lambda_i|}$, where $\lambda_i$ is the eigenvalues of the matrix. Thus, combining $(1)$ and $(2)$ results in

$$\|A\|_2=\sigma_1(A)=\max_i |\lambda_i|=\rho(A) \quad \quad â–  $$

The Gershgorin circle theorem states that the closed disc with radius $R_i$ and centered at $a_{ii}$ for row $i$ is given by $R_i=\sum_{j\neq i} |a_{ij}|=1 \forall i\neq 1$. According to the theorem, all eigenvalues, ($0$), must be located within at least one of these discs. The origo is within all discs, so the theorem holds.

**(c)** By definition, $$K_2(\hat{A})=\|\hat{A}\|_2\cdot \|\hat{A}^{-1}\|_2=\sqrt{\rho\left(\hat{A}^H\hat{A}\right)\rho\left(\left(\hat{A}^H\hat{A}\right)^{-1}\right)}$$

What remains is to find $\hat{A}^H\hat{A}$ and $\left(\hat{A}^H\hat{A}\right)^{-1}$, respectively, and find their largest eigenvalues.

$$
\hat{A}^H=\left[
\begin{array}{rrrr}
0 & 1 & \ldots & 0  \\
0 & 0 & \ddots  & \vdots \\
\vdots & \vdots &\ddots & 1 \\
\varepsilon & 0 & \ldots & 0 &

\begin{array}{rrrr}

0 & 0 & \ldots & \varepsilon^2 & 
\end{array}
\right]
$$
Which is a diagonal matrix. Thus,
$$\lambda(\hat{A}^H\hat{A})={1, 1, \dots, \varepsilon^2}\\
\implies \rho(\hat{A}^H\hat{A})=\max \{1, |\varepsilon^2|\}=1,
$$
assuming $0<\varepsilon<1$.

Findind the inverse is relatively easy as well, since the only operation needed is to divide the last row by $\varepsilon^2$. Thus,

$$
\left(\hat{A}^H\hat{A}\right)^{-1}=\left[
\begin{array}{rrrr}
1 & 0 & \ldots & 0  \\
0 & 1 & \ldots  & 0 \\
\vdots &  &\ddots & \vdots \\
0 & 0 & \ldots & \frac{1}{\varepsilon^2} &

\right],
$$
which is also a diagonal matrix. Hence,
$$\lambda\left(\hat{A}^H\hat{A}\right)^{-1}= {1, 1, \dots, \frac{1}{\varepsilon^2} }\\
\implies \rho\left(\left(\hat{A}^H\hat{A}\right)^{-1}\right)=\frac{1}{\varepsilon^2},$$
assuming $0<\varepsilon<1$.

$$\therefore K_2=\sqrt{1\cdot \varepsilon^{-2}}=\frac{1}{\varepsilon}$$

<font color='grey'>The Hilbert matrix is a classical example of an ill-conditioned matrix. For an $n\times n$-matrix it is

**(a)**
For $n=2,\ldots,13$ find the $n\times n$ Hilbert matrix $H_n$ and compute the condition number $K_2(H_n)$ in Python. You can either write a function yourself for setting up $H_n$, or you can use scipy.linalg.hilbert().
To compute the condition number you can use numpy.linalg.cond. Display/plot the result in a way that best shows the growth rate of the condition number as a function of the dimension $n$.

**(b)** Illustrate the meaning of condition numbers as follows: For a range of small perturbations $\delta H_n$ and
$\delta b$, solve the problems $(H_n+\delta H_n)(x+\delta x) = b+\delta b$ where $b=[1,\ldots,1]^T$ (the $n$-vector of ones). In each case, compare to the solution of $H_n x = b$ and calculate

$$
     \frac{\frac{\|\delta x\|}{\|x\|}}{\frac{\|\delta H_n\|}{\|H_n\|}+\frac{\|\delta b\|}{\|b\|}}.
$$

One way of generating random arrays (matrices and vectors) is to use numpy.random.rand(j,k). This generates a $j \times k$ array, where each entry is a random number in the range $[0,1).$ Remember however that the smallest entry in $H_n$ is $\tfrac{1}{2n+1}$ so yopu have to make sure that the matrix entries in $\delta H_n$ are small enough to consider $\delta H_n$ a small perturbation.</font>

```python
## PROBLEM 3 A ## 

import scipy.linalg as sp
import numpy as np
import matplotlib.pyplot as plt

def computeCondNumHilbert(n1, n2):
    K2s = np.zeros(n2-n1+1) #container for condition numbers
    for n in range(n1,n2+1):
        tHilbert = sp.hilbert(n)
        tK2 = np.linalg.cond(tHilbert)
        K2s[n-n1] = tK2

    #Plotting
    plt.yscale('log')
    xax = np.linspace(n1,n2,n2-n1+1) #Shifts x-axis so that K2s[0] plots to n=2
    plt.plot(xax, K2s, label="$K_2(n)$ of dimension $n$")
    plt.xlabel("n")
    plt.ylabel("$K_2(n)$")
    plt.legend()


computeCondNumHilbert(2,13)
plt.show()
```

```python
## PROBLEM 3B ##
def smolPertubations(H, b, scale):
    n = H.shape[0]
    dH = scale*1/(2*n+1)*np.random.rand(n,n) #scale random changes about 1% the size of H and b
    db = scale*np.random.rand(n)
    return dH, db
def illustrateCondNo(n1, n2):
    normRels = [] #containers to store values for plotting
    xax = []
    for n in range(n1, n2+1):
        H = sp.hilbert(n)
        b = np.ones(n)
        dH, db = smolPertubations(H, b, 1e-15)
        #By trial and error, scale had to be about 1e-15 for perturbations to be considered small
        x_dx=np.linalg.solve(H+dH,b+db) #Solves (x+dx)
        x = np.linalg.solve(H,b) #Solves (x)
        dx = x_dx-x #Finds dx
        #Calculates norm ratio:
        x_rel = np.linalg.norm(dx)/np.linalg.norm(x)
        H_rel = np.linalg.norm(dH)/np.linalg.norm(H)
        b_rel = np.linalg.norm(db)/np.linalg.norm(b)
        rel = x_rel / (H_rel + b_rel)
        normRels.append(rel)
        xax.append(n)
    plt.yscale('log')
    plt.plot(xax, normRels, label="Norm relationship", color="orange")
    plt.legend()

illustrateCondNo(2,13)
computeCondNumHilbert(2,13) #The function from 2a for reference
plt.show()
print("The graph illustrates how the norm relationship is a lower bound for the condition number")
```
